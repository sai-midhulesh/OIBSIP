```python
def dummies(x,df):
    temp = pd.get_dummies(df[x],drop_first=True)
    df = pd.concat([df,temp],axis=1)
    df.drop(x,axis=1,inplace=True)
    return df
```

```python

df = dummies('company_name',df)
df = dummies('doornumber',df)
df = dummies('enginelocation',df)
df = dummies('fuelsystem',df)

```
